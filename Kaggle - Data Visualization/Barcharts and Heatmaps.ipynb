```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('flight_delays.csv', index_col="Month")
```

```python
df
```

## Box plot

```python
plt.figure(figsize=(10,6))

plt.title("Average Arrival Delay")
sns.barplot(x=df.index, y=df['AA'])
plt.ylabel("Arrival delay (in minutes)")
```

## Heatmap

```python
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=df, annot=True) # annot ensured that the values for each cell appear on the chart
plt.xlabel('Airline')
```

## Scatter Plots

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x=df['age'], y=df['charges'])
```

```python
# body mass index (bmi) and insurance charges are positively correlated => customers with higher bmi 
# tipically also tend to pay more in insurance costs.
sns.scatterplot(x=df['bmi'], y=df['charges'])
```

```python
# regplot shows the regression line that best fits the data
sns.regplot(x=df['bmi'], y=df['charges'])
```

```python
#lmplot adds two regression lines, corresponding to smokers and nosmokers
sns.lmplot(x="bmi", y="charges", hue="smoker", data=df)
```

```python
sns.lmplot(x="bmi", y="charges", hue="sex", data=df)
```

```python
#categorical scatter plot
# Shows the relationship between a categorical variable (smoker) and a continuous variable('charges)')
# The plot shows that: on average, non-smokers are charged less than smokers, and the customers
# who pay the least are non-smokers
sns.swarmplot(x=df['smoker'], y=df['charges'])
```

## Distributions

```python
df = pd.read_csv('iris.csv')
```

###### Histograms

```python
sns.distplot(a=df['Petal Length (cm)'])
```

###### Density plots

```python
sns.kdeplot(data=df['Petal Length (cm)'], shade=True)
```

```python
# Multiple plots 

iris_set = df[df.Species == 'Iris-setosa']
iris_ver = df[df.Species == 'Iris-versicolor']
iris_virg = df[df.Species ==  'Iris-virginica']

sns.distplot(a=iris_set['Petal Length (cm)'], label='Iris-setosa')
sns.distplot(a=iris_ver['Petal Length (cm)'], label='Iris-versicolor')
sns.distplot(a=iris_virg['Petal Length (cm)'], label='Iris-virginica')
```

```python
sns.kdeplot(data=iris_set['Petal Length (cm)'], label='Iris-setosa',shade=True)
sns.kdeplot(data=iris_ver['Petal Length (cm)'], label='Iris-versicolor', shade=True)
sns.kdeplot(data=iris_virg['Petal Length (cm)'], label='Iris-virginica', shade=True)
```
